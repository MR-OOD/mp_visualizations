```python
import os
from PIL import Image
from collections import defaultdict
```

```python
# --- CONFIG ---
input_dir = r"C:\Users\justi\OneDrive\Desktop\University\Master_Studium\MP\code\sCT-ood\results\ReverseDistillationCustomLR\rd4ad_radimagenet_001\test_images"
output_dir = r"C:\Users\justi\OneDrive\Desktop\University\Master_Studium\MP\code\sCT-ood\results\ReverseDistillationCustomLR\rd4ad_radimagenet_001\gifs"

os.makedirs(output_dir, exist_ok=True)
duration = 250  # milliseconds per frame in the GIF
loop = 0        # 0 = infinite loop

os.makedirs(output_dir, exist_ok=True)

# --- Group images by patient ID ---
patients = defaultdict(list)
for fname in os.listdir(input_dir):
    if fname.endswith(".png"):
        patient_id = fname.split("_")[0]
        patients[patient_id].append(fname)

# --- Create GIF for each patient ---
for patient_id, files in patients.items():
    files.sort(key=lambda x: int(x.split("_")[1].split(".")[0]))  # sort by slice number
    images = [Image.open(os.path.join(input_dir, f)) for f in files]
    
    gif_path = os.path.join(output_dir, f"{patient_id}.gif")
    images[0].save(
        gif_path,
        save_all=True,
        append_images=images[1:],
        duration=duration,
        loop=loop
    )
    print(f"✅ Saved {gif_path}")

print("All GIFs created!")
```
